```python
uniq_users=filtered_df['PNR'].unique()
```

```python
len(uniq_users)
```

```python
# list1=[]
pnr=[]
remark=[]
travel_his=[]
home_code=[]
ori = []
des = []
for i in range(0, len(uniq_users)):
    
  s = []
  # print(i)
  user = filtered_df[filtered_df['PNR']==uniq_users[i]].sort_values(by=['Dates_dept','Time_dept','Time_arr'])
  # print(user)
  user_orig_des = user[['PNR','orig_city_cd', 'dest_city_cd']]
  user_orig_des['combined'] = user_orig_des['orig_city_cd'].astype(str) + '->' + user_orig_des['dest_city_cd']
  
  for index, row in user_orig_des.iterrows():
      x = row["orig_city_cd"]
      y = row["dest_city_cd"]
      s.extend([x,y])
  print(s)
  travel_his.append(s)
  pnr.append( uniq_users[i])
      # break
  ori.append(s[0])
  des.append(s[-1])
    
  if s[0]==s[-1]:
    print("Round trip with origin city as:",s[0])
    remark.append('Round trip')
    home_code.append("PHX")
 

  else:
    print("One way trip with origin city as:",s[0])
    remark.append('one way trip')
    home_code.append("PHX")
  

```

```python
COLUMN_NAMES=['PNR','travel_history', 'ori','des','Trip Type','home_code']
new_df=pd.DataFrame(columns=COLUMN_NAMES)
```

```python
new_df['PNR'] = pnr
new_df['travel_history'] = travel_his
new_df["ori"] = ori
new_df['des'] = des
new_df['Trip Type']=remark
new_df['home_code'] = home_code

```
